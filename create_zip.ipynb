```python
import os
import shutil
from zipfile import ZipFile

# Define your source directories and number of zip files to create
source_folders = ['DDR-dataset/DR_grading/imgs_with_classes/0', 'DDR-dataset/DR_grading/imgs_with_classes/1', 'DDR-dataset/DR_grading/imgs_with_classes/2', 'DDR-dataset/DR_grading/imgs_with_classes/3', 'DDR-dataset/DR_grading/imgs_with_classes/4']
destination_root_base = 'DDR-dataset/DR_grading/zip_files'
num_zip_files = 47  # Number of zip files to create
labels = {'0': 'No DR', '1': 'Mild DR', '2': 'Moderate DR', '3': 'Severe DR', '4': 'Proliferative DR' }

for zip_index in range(num_zip_files):
    # Create a unique destination root for each zip file
    destination_root = f'{destination_root_base}/set_{zip_index + 1}'
    if not os.path.exists(destination_root):
        os.makedirs(destination_root)

    # Loop through each source folder
    for folder in source_folders:
        source_path = os.path.join('.', folder)
        files = sorted([f for f in os.listdir(source_path) if os.path.isfile(os.path.join(source_path, f))])
        start_index = zip_index * 5
        selected_files = files[start_index:start_index + 5]
        
        # Create a destination folder for the selected images
        folder_name = labels[os.path.basename(folder)]
        destination_folder = os.path.join(destination_root, f'{folder_name}')
        if not os.path.exists(destination_folder):
            os.makedirs(destination_folder)
        
        # Copy the selected images to the new destination
        for file in selected_files:
            shutil.copy(os.path.join(source_path, file), destination_folder)

    # Zip the created folders
    zip_filename = f'{destination_root_base}/set_{zip_index + 1}.zip'
    with ZipFile(zip_filename, 'w') as zipf:
        for root, dirs, files in os.walk(destination_root):
            for file in files:
                zipf.write(os.path.join(root, file), os.path.relpath(os.path.join(root, file), os.path.join(destination_root, '..')))

    print(f'Created zip file: {zip_filename}')
```
